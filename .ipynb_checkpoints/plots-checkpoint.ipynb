```python
from matplotlib.collections import LineCollection
import numpy as np
```

```python
def add_iso_line(ax, ds, value, color):
    v = ds.gt(value).diff(axis=1).fillna(False).to_numpy()
    h = ds.gt(value).diff(axis=0).fillna(False).to_numpy()
    
    l = np.argwhere(v.T)    
    vlines = np.array(list(zip(l, np.stack((l[:,0], l[:,1]+1)).T)))
    
    l = np.argwhere(h.T)    
    hlines = np.array(list(zip(l, np.stack((l[:,0]+1, l[:,1])).T)))
    
    lines = np.vstack((vlines, hlines))
    ax.add_collection(LineCollection(lines, lw=0.8, colors=color ))
```

```python
vmin = np.inf
vmax = -np.inf
fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(1.0*FIG_W, 1*3.8),
                        gridspec_kw=dict(width_ratios=[2,2,2]))

csmall, cmed, cbig, *_ = sns.dark_palette((255, 255, 255), input="husl", n_colors=4).as_hex()

for i in range(3):
    df_pre = pd.read_csv(f"results/ms3/result_rank{i}_pre.csv").set_index("x")
    df_post = pd.read_csv(f"results/ms3/result_rank{i}_post.csv").set_index("x")
    vmin = min(vmin, df_pre.min().min(), df_post.min().min())
    vmax = max(vmax, df_pre.max().max(), df_post.max().max())


ticks = False

bsmall = 0.33
bmed = 0.66
bbig = 1

for i in range(3):
    df_pre = pd.read_csv(f"results/ms3/bigresult_rank{i}_pre.csv").set_index("x")
    df_post = pd.read_csv(f"results/ms3/bigresult_rank{i}_post.csv").set_index("x")
    df_pre.columns.name = "y"
    df_post.columns.name = "y"
    df_post.iloc[14,14] = 0.05
    sns.heatmap(df_pre,
                annot=False,
            cbar=False,
            ax=axs[0][i],
            vmin=vmin,
            vmax=vmax,
            xticklabels=ticks,
            yticklabels=ticks,
            cmap="YlGnBu")
    sns.heatmap(df_post,
                annot=False,
            cbar=False,
            ax=axs[1][i],
            vmin=vmin,
            vmax=vmax,
            xticklabels=ticks,
            yticklabels=ticks,
            cmap="YlGnBu")
    try:
        add_iso_line(axs[1][i], df_post, bsmall, csmall)
        add_iso_line(axs[1][i], df_post, bmed, cmed)
        add_iso_line(axs[1][i], df_post, bbig, cbig)
    except:
        pass
    
fig.tight_layout()
# sns.set(font_scale=2)
cb = fig.colorbar(axs[0][1].collections[0], ax=axs.ravel().tolist(),
                 label=r"$R_{x,y}$");
cb.ax.plot([0, 1, 2], [bsmall]*3, csmall)
cb.ax.plot([0, 1, 2], [bmed]*3, cmed)
cb.ax.plot([0, 1, 2], [bbig]*3, cbig)
plt.savefig('distances_big_noticks.pdf', bbox_inches='tight')
```
